```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

path to the csv file on github

```python
url = 'https://raw.githubusercontent.com/dbecher84/IS362_assignment8/master/auto_mpg.data.csv'
```

read the csv file into a pandas dataframe

```python
car_data = pd.read_csv(url)
```

add column headers to data frame from the auto-mpg.names file

```python
car_data.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
```

replace the '?' with NaN and change the horsepower column from object to numeric

```python
car_data.replace('?', np.nan, inplace=True)
car_data['horsepower'] = car_data['horsepower'].apply(pd.to_numeric)
```

change the origin column 
1 = USA,  2 = Asia,  3 = Europe

```python
car_data['origin'] = car_data['origin'].map({1:'USA', 2:'Asia', 3:'Europe'})
```

make the car name the index value

```python
car_data.set_index('car name', inplace=True)
```

```python
car_data[:40]
```

Bar chart showing distribution of cylinders

```python
sns.countplot(x='cylinders', data=car_data)
```

Scatter plot showing horsepower and weight

```python
sns.lmplot(x='weight', y='horsepower', data=car_data, fit_reg=True)
```

The USA seems to have more cars with higher horsepower and a larger engine diplacement.
